```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('train.csv')

df1.head(10)
```

```python
df_ = pd.read_csv('test.csv')
df_.head(10)
```

```python
df__ = pd.read_csv('gender_submission.csv')
df__.head(10)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
X_test_tbm = df_.set_index('PassengerId')[['Age' ,'SibSp' ,'Parch' ,'Fare'  ]]
df11 = pd.get_dummies(df_.set_index('PassengerId')[['Sex' , 'Embarked']],columns = ['Sex','Embarked'] ,drop_first =True )
X_test = pd.concat([X_test_tbm , df11 ] ,axis =1)
X_test['Age'].fillna(np.nanmedian(X_test['Age']) , inplace =True)
X_test['Fare'].fillna(np.nanmedian(X_test['Fare']) , inplace =True)
X_test.head(10)
```

```python
X_tbm = df1.set_index('PassengerId')[['Age' ,'SibSp' ,'Parch' ,'Fare'  ]]
X_tbm.head(10)
```

```python
df00 = pd.get_dummies(df1.set_index('PassengerId')[['Sex' , 'Embarked']],columns = ['Sex','Embarked'] ,drop_first =True )
X = pd.concat([X_tbm , df00 ] ,axis =1)
X.head(10)
```

```python
y = df1[['PassengerId','Survived']].set_index('PassengerId')
y.head(10)
```

```python
Xy = pd.concat([X,y] ,axis =1)
Xy.shape
```

```python
Xy.dropna(inplace =True)
Xy.shape
```

```python
Xy.head(10)
```

```python
X_full = Xy[['Age' ,'SibSp', 'Parch', 'Fare', 'Sex_male', 'Embarked_Q', 'Embarked_S' ]]
y_full = Xy[['Survived']]
X_full.head(10)
```

```python
y_full.head(10)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
```

```python
a = cross_val_score(RandomForestClassifier() , X_full ,y_full ,cv =10)
(np.mean(a) ,np.var(a) ,np.std(a))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model1 = GridSearchCV(RandomForestClassifier(random_state =0) , param_grid = {'max_features' : [1,2,3] , 'max_depth' : [2,3,4,5,6,7]} ,cv =10)
```

```python
model1.fit(X_full ,y_full)
```

```python
model_final = model1.best_estimator_
```

```python
model_final.score(X_full ,y_full)
```

```python
lst = model_final.predict(X_test)
df= pd.DataFrame(lst , index = X_test.index)
df.reset_index(inplace =True)
df.rename({0 :'Survived' } , axis = 'columns' , inplace =True)
df
```

```python
df.to_csv('Survived_final.csv' ,index = False)
```
